```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import joblib

# Dummy dataset
df = pd.DataFrame({
    'age': np.random.randint(22, 60, 100),
    'education': np.random.choice(['Bachelors', 'Masters', 'PhD', 'HS-grad'], 100),
    'occupation': np.random.choice(['Engineer', 'Manager', 'Technician', 'Clerk'], 100),
    'experience': np.random.randint(0, 35, 100),
    'hours-per-week': np.random.randint(20, 60, 100),
    'salary': np.random.randint(25000, 150000, 100)
})

X = df.drop('salary', axis=1)
y = df['salary']

# Preprocessing pipeline
categorical = ['education', 'occupation']
preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical)
], remainder='passthrough')

model = Pipeline([
    ('prep', preprocessor),
    ('reg', LinearRegression())
])

model.fit(X, y)
joblib.dump(model, 'best_salary_model.pkl')
```

```python
# app.py

import streamlit as st
import pandas as pd
import numpy as np
import joblib
from streamlit_lottie import st_lottie
import json

# --- Load Animation ---
def load_lottie_file(filepath: str):
    with open(filepath, "r") as f:
        return json.load(f)

animation = load_lottie_file("animation.json")

# --- Page Config ---
st.set_page_config(page_title="Employee Salary Predictor", layout="wide")

# --- Header ---
with st.container():
    st.title("💼 Employee Salary Prediction App")
    st_lottie(animation, height=200, key="salary_pred_anim")
    st.write("### Predict employee salary based on work experience and skills.")

# --- Sidebar ---
with st.sidebar:
    st.header("🧾 Input Features")
    experience = st.slider("Years of Experience", 0, 30, 5)
    education_level = st.selectbox("Education Level", ["High School", "Bachelor's", "Master's", "PhD"])
    role = st.selectbox("Job Role", ["Data Analyst", "Software Engineer", "Manager", "HR", "Accountant"])

# --- Feature Engineering ---
edu_map = {"High School": 0, "Bachelor's": 1, "Master's": 2, "PhD": 3}
role_map = {"Data Analyst": 0, "Software Engineer": 1, "Manager": 2, "HR": 3, "Accountant": 4}

input_df = pd.DataFrame({
    "experience": [experience],
    "education_level": [edu_map[education_level]],
    "role": [role_map[role]]
})

# --- Load or Create Dummy Model ---
try:
    model = joblib.load("salary_model.pkl")
except:
    from sklearn.linear_model import LinearRegression
    X_dummy = pd.DataFrame({
        "experience": np.random.randint(0, 30, 100),
        "education_level": np.random.randint(0, 4, 100),
        "role": np.random.randint(0, 5, 100)
    })
    y_dummy = 20000 + X_dummy["experience"]*1500 + X_dummy["education_level"]*3000 + X_dummy["role"]*2000
    model = LinearRegression().fit(X_dummy, y_dummy)
    joblib.dump(model, "salary_model.pkl")

# --- Prediction ---
prediction = model.predict(input_df)[0]

st.subheader("📊 Prediction Result")
st.success(f"Estimated Salary: ₹{int(prediction):,}")
```

```python
import streamlit as st
import pandas as pd
import joblib
import json
import os
from streamlit_lottie import st_lottie

# Load Lottie animation
def load_lottie_file(filepath: str):
    if os.path.exists(filepath):
        with open(filepath, "r") as f:
            return json.load(f)
    return None

animation = load_lottie_file("animation.json")

# Load the model
model = joblib.load("best_model.pkl")  # Make sure this file exists

# Streamlit UI
st.set_page_config(page_title="Salary Predictor", page_icon="💼", layout="wide")

# Sidebar
with st.sidebar:
    st.title("💼 Salary Prediction")
    st.markdown("Use the form below to enter employee details.")

    age = st.slider("Age", 18, 65, 30)
    gender = st.selectbox("Gender", ["Male", "Female"])
    education = st.selectbox("Education Level", ["High School", "Bachelor", "Master", "PhD"])
    experience = st.slider("Years of Experience", 0, 40, 5)
    job_role = st.selectbox("Job Role", ["Developer", "Manager", "Analyst", "HR", "Accountant"])

# Animation or heading
with st.container():
    st.title("Employee Salary Prediction 💰")
    if animation:
        st_lottie(animation, height=200, key="salary_pred_anim")
    else:
        st.info("No animation available.")

# Map input to model format
input_data = pd.DataFrame({
    "age": [age],
    "gender": [1 if gender == "Male" else 0],
    "education": [education],
    "experience": [experience],
    "job_role": [job_role]
})

# One-hot encoding for categorical columns (should match training preprocessing)
input_data = pd.get_dummies(input_data)

# Match input columns to model
expected_cols = model.feature_names_in_
for col in expected_cols:
    if col not in input_data.columns:
        input_data[col] = 0
input_data = input_data[expected_cols]

# Predict
if st.button("🔮 Predict Salary"):
    prediction = model.predict(input_data)[0]
    st.success(f"💸 Predicted Salary: ₹ {int(prediction):,} per annum")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
from streamlit_lottie import st_lottie
import json
import joblib

# Load Lottie animation
def load_lottie_file(filepath: str):
    with open(filepath, "r") as f:
        return json.load(f)

animation = load_lottie_file("animation.json")

# Page config
st.set_page_config(page_title="Salary Predictor", page_icon="💼", layout="centered")
st_lottie(animation, height=180, key="salary_pred_anim")

st.title("💼 Employee Salary Prediction App")

# User inputs
st.subheader("Enter Employee Details:")
experience = st.slider("Years of Experience", 0, 20, 3)
education = st.selectbox("Education Level", ["High School", "Bachelor's", "Master's", "PhD"])
department = st.selectbox("Department", ["HR", "Engineering", "Sales", "Marketing", "Finance"])
location = st.selectbox("Location", ["Urban", "Suburban", "Rural"])

# Encode inputs (simple encoding for demo)
education_map = {"High School": 0, "Bachelor's": 1, "Master's": 2, "PhD": 3}
dept_map = {"HR": 0, "Engineering": 1, "Sales": 2, "Marketing": 3, "Finance": 4}
location_map = {"Urban": 2, "Suburban": 1, "Rural": 0}

input_data = np.array([
    experience,
    education_map[education],
    dept_map[department],
    location_map[location]
]).reshape(1, -1)

# Load trained model (dummy logic here for example)
def dummy_model(data):
    return 30000 + data[0][0]*4000 + data[0][1]*5000 + data[0][2]*1000 + data[0][3]*2000

if st.button("Predict Salary"):
    salary = dummy_model(input_data)
    st.success(f"Estimated Salary: ₹ {int(salary):,}/year")
```
